```python
#Conversion of mp3 to wav File
path= '/home/fractaluser/Desktop/des/data/test/'
from pydub import AudioSegment
import os
from os import listdir
from os.path import isfile, join

AudioSegment.converter = r"ffmpeg"#r"/usr/local/lib/python2.7/dist-packages/tensorflow/contrib/ffmpeg"
#change the range(10)
for i in range(2):
    files = os.listdir(path+r'''audio{}'''.format(i))
    
    if not os.path.exists(path+r'''wav{}'''.format(i)):
        os.makedirs(path+r'''wav{}'''.format(i))
    
    count = 1
    limit = 0
    for f in files:
        if(limit == 720): break
        sound = AudioSegment.from_mp3(path+r'''audio{}/{}'''.format(i, f))
        out_f = sound.export(path+r'''wav{}/{}.wav'''.format(i, count), format="wav")
        out_f.close()
        count = count + 1
        limit = limit + 1
```

```python
#Cutting down the .wav file(multiple wav files per directory) to 5 seconds interval
import subprocess
import re
import os
count=0
notfile=5
#change the range(10)
for i in range(2):
    in_path = os.listdir(path+'wav{}'.format(i))
    if not os.path.exists(path+r'''wavs{}'''.format(i)):
        os.makedirs(path+r'''wavs{}'''.format(i))    
    for z in in_path:
        r = path+'wav{}/'.format(i)+z  
        out_path = path+'wavs{}/out%03d{}'.format(i,z)
        process = subprocess.Popen(['ffmpeg',  '-i', r], stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
        stdout, stderr = process.communicate()
        matches = re.search(r"Duration:\s{1}(?P<hours>\d+?):(?P<minutes>\d+?):(?P<seconds>\d+\.\d+?),", stdout, re.DOTALL).groupdict()
            # print matches['hours']
        c=matches['minutes']
        a=matches['seconds']
        b=float(a)
        c= int(c)
        totaltime=c*60+int(b)
        print(totaltime)
        start=0
        count=0
        z=0
        end=5
        while(z<=totaltime):             
            cmdstring='ffmpeg -i "{}" -ss "{}" -t "{}" -acodec copy "{}"'.format(r,r'''{}'''.format(start),r'''{}'''.format(notfile), path+r'''wavs{}/output{}.wav'''.format(i, count))
            subprocess.call(cmdstring, shell=True)
#                 print(cmdstring)
            count=count+1
            start=end-2
            end=start+5
            z=end
```
